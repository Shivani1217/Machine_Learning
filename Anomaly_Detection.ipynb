1.  What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.
- Anomaly detection is a technique used to identify data points or patterns that deviate significantly from normal behavior. These unusual patterns are called anomalies or outliers. The goal is to detect events that do not follow expected trends, since they often indicate errors, fraud, faults, or rare but important situations.

In data science and machine learning, anomaly detection is widely used in areas like fraud detection, network security, healthcare monitoring, and system performance analysis.

### Types of Anomalies

**1. Point Anomalies**
A point anomaly occurs when a single data point is significantly different from the rest of the data.

**Example:**

* In a dataset of daily bank transactions where most transactions are between ₹500 and ₹10,000, a transaction of ₹5,00,000 would be a point anomaly.
* In temperature sensor data where readings are usually around 25–30°C, a sudden reading of 80°C is a point anomaly.

This is the simplest and most common type of anomaly.

**2. Contextual Anomalies**
A contextual anomaly is data that is abnormal only in a specific context. The context could be time, location, season, or any other condition.

**Example:**

* A temperature of 35°C is normal in summer but would be anomalous in winter.
* High electricity usage at noon may be normal for a factory, but the same usage at midnight could be considered abnormal.

Here, the data point is not inherently unusual, but it becomes an anomaly because of the surrounding context.

**3. Collective Anomalies**
A collective anomaly occurs when a group of related data points together form an abnormal pattern, even if individual points look normal on their own.

**Example:**

* A sudden sequence of small but repeated login attempts may indicate a brute-force attack, even though each attempt looks normal individually.
* In ECG data, a series of slightly irregular heartbeats together may indicate a heart condition, even if a single heartbeat does not look abnormal.

In this case, the anomaly is detected in the pattern of values rather than individual points.




2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.
- Here’s a clear, exam- and interview-friendly comparison of **Isolation Forest**, **DBSCAN**, and **Local Outlier Factor (LOF)**, focusing on their approach and when to use each.

---

### 1. Isolation Forest (iForest)

**Approach**
Isolation Forest is a tree-based anomaly detection algorithm.
Instead of modeling normal data, it isolates anomalies by randomly selecting features and split values. Since anomalies are rare and different, they get isolated in fewer splits, resulting in shorter tree paths.

**Key idea:**
Anomalies are easier to isolate than normal points.

**Suitable Use Cases**

* Large and high-dimensional datasets
* Fraud detection (banking, insurance)
* Network intrusion detection
* Situations where labels are not available

**Strengths**

* Scales well to big data
* Works well in high dimensions
* Fast and memory efficient

**Limitations**

* Less interpretable than distance-based methods
* May not capture local density variations well

---

### 2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

**Approach**
DBSCAN is a density-based clustering algorithm.
It groups points that are closely packed together and marks points in low-density regions as noise (outliers).

**Key idea:**
Points far from dense clusters are anomalies.

**Suitable Use Cases**

* Spatial or geographical data
* Datasets with clusters of arbitrary shapes
* Noise detection in clustering tasks
* Small to medium-sized datasets

**Strengths**

* Does not require the number of clusters in advance
* Can detect arbitrary-shaped clusters
* Naturally identifies noise points

**Limitations**

* Sensitive to parameters (epsilon and minPts)
* Struggles with varying densities
* Not suitable for high-dimensional data

---

### 3. Local Outlier Factor (LOF)

**Approach**
LOF is a density-based anomaly detection method that compares the local density of a point with the densities of its neighbors.
A point is considered an anomaly if it has significantly lower density than its surrounding points.

**Key idea:**
Anomalies have lower local density compared to their neighbors.

**Suitable Use Cases**

* Detecting local anomalies in complex datasets
* Data with varying densities
* Fraud detection and sensor data analysis

**Strengths**

* Effective for detecting local outliers
* Handles varying density regions better than DBSCAN
* Good for nuanced anomaly patterns

**Limitations**

* Computationally expensive for large datasets
* Sensitive to choice of number of neighbors (k)
* Not ideal for very high-dimensional data

---

### Quick Comparison Table

| Algorithm        | Approach                     | Best For                      | Main Limitation                 |
| ---------------- | ---------------------------- | ----------------------------- | ------------------------------- |
| Isolation Forest | Random isolation using trees | Large, high-dimensional data  | Weak at local density anomalies |
| DBSCAN           | Density-based clustering     | Spatial data, noise detection | Sensitive to parameters         |
| LOF              | Local density comparison     | Local and contextual outliers | Computationally expensive       |

3.  What are the key components of a Time Series? Explain each with one example.
- A **time series** is a sequence of data points collected or recorded at regular time intervals. Time series analysis focuses on understanding the underlying patterns in this data over time.

The key components of a time series are **Trend, Seasonality, Cyclical, and Irregular (Random) components**.

---

### 1. Trend

The **trend** represents the long-term direction or movement in the data over a period of time. It shows whether the data is generally increasing, decreasing, or stable.

**Example:**

* The steady increase in mobile phone users in India over the last 10 years shows an upward trend.

---

### 2. Seasonality

**Seasonality** refers to regular and repeating patterns that occur at fixed intervals due to seasonal factors like weather, festivals, or calendar effects.

**Example:**

* Ice cream sales increase every summer and decrease during winter each year.

---

### 3. Cyclical Component

The **cyclical component** represents long-term fluctuations that occur over periods longer than one season. These cycles are usually influenced by economic or business conditions and do not have a fixed length.

**Example:**

* An economy experiencing periods of growth and recession over several years shows cyclical behavior.

---

### 4. Irregular (Random) Component

The **irregular component** includes unpredictable and random variations that cannot be explained by trend, seasonality, or cycles.

**Example:**

* A sudden drop in stock prices due to unexpected political news.




4.  Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?
- ### Stationarity in Time Series

A **time series is said to be stationary** if its statistical properties remain constant over time. This means the series has:

* Constant mean
* Constant variance
* Constant autocovariance (relationship between values depends only on the lag, not on time)

Stationarity is important because most time series models like ARIMA assume the data is stationary.

---

### How to Test Stationarity

**1. Visual Inspection**

* Plot the time series and check for trends or changing variance.
* A trending or seasonal pattern usually indicates non-stationarity.

**2. Rolling Statistics**

* Calculate rolling mean and rolling variance.
* If they change over time, the series is likely non-stationary.

**3. Augmented Dickey-Fuller (ADF) Test**

* **Null hypothesis (H₀):** The series has a unit root (non-stationary).
* If the p-value is less than 0.05, reject H₀ and conclude the series is stationary.

**4. KPSS Test**

* **Null hypothesis (H₀):** The series is stationary.
* A low p-value suggests the series is non-stationary.

Using both ADF and KPSS together gives a more reliable conclusion.

---

### How to Transform a Non-Stationary Series into a Stationary One

**1. Differencing**

* Subtract the previous value from the current value.
* Removes trend and stabilizes the mean.

**Example:**
First-order differencing:
( Y'*t = Y_t - Y*{t-1} )

**2. Log or Power Transformation**

* Apply log, square root, or Box-Cox transformation.
* Helps stabilize variance when variance increases over time.

**Example:**
Log transformation:
( Y'_t = \log(Y_t) )

**3. Seasonal Differencing**

* Subtract values from the same season in the previous cycle.
* Used to remove seasonality.

**Example:**
Monthly data with yearly seasonality:
( Y'*t = Y_t - Y*{t-12} )

**4. Detrending**

* Remove the trend component using regression or moving averages.

**5. Smoothing (less common for modeling)**

* Moving averages can reduce noise but are mainly used for visualization.


5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.
- Below is a clear differentiation of **AR, MA, ARIMA, SARIMA, and SARIMAX** models, focusing on their **structure** and **applications**. This format works well for exams, interviews, and practical understanding.

---

### 1. AR (Autoregressive) Model

**Structure**

* Uses past values of the series to predict the current value.
* Depends on **p** lagged observations.

[
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \dots + \phi_p Y_{t-p} + \epsilon_t
]

**Application**

* Used when the series is stationary and shows strong correlation with its past values.
* Examples: stock returns, temperature readings.

---

### 2. MA (Moving Average) Model

**Structure**

* Uses past forecast errors (residuals) to model the current value.
* Depends on **q** lagged error terms.

[
Y_t = c + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q} + \epsilon_t
]

**Application**

* Useful when shocks or random fluctuations affect the series.
* Examples: sensor noise, short-term economic shocks.

---

### 3. ARIMA (Autoregressive Integrated Moving Average)

**Structure**

* Combines AR and MA with **d** order differencing to handle non-stationary data.
* Written as ARIMA(p, d, q).

[
\text{ARIMA}(p,d,q)
]

**Application**

* Used for non-seasonal time series that become stationary after differencing.
* Examples: sales forecasting, demand prediction.

---

### 4. SARIMA (Seasonal ARIMA)

**Structure**

* Extends ARIMA by adding seasonal components.
* Written as SARIMA(p, d, q)(P, D, Q, s), where **s** is the seasonal period.

[
\text{SARIMA}(p,d,q)(P,D,Q)_s
]

**Application**

* Used when data shows repeating seasonal patterns.
* Examples: monthly sales, electricity consumption.

---

### 5. SARIMAX (Seasonal ARIMA with eXogenous variables)

**Structure**

* SARIMA model plus external (exogenous) variables that influence the series.
* Incorporates additional predictors along with time-based components.

**Application**

* Used when external factors affect the target variable.
* Examples: sales influenced by advertising spend, temperature affecting energy usage.

---

### Quick Comparison Table

| Model   | Structure                 | Seasonality | External Variables | Typical Use                       |
| ------- | ------------------------- | ----------- | ------------------ | --------------------------------- |
| AR      | Past values only          | No          | No                 | Stationary data                   |
| MA      | Past errors only          | No          | No                 | Noise-driven series               |
| ARIMA   | AR + MA + differencing    | No          | No                 | Non-seasonal forecasting          |
| SARIMA  | ARIMA + seasonal terms    | Yes         | No                 | Seasonal time series              |
| SARIMAX | SARIMA + exogenous inputs | Yes         | Yes                | Forecasting with external factors |

7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.


```python
# Isolation Forest for Anomaly Detection (NYC Taxi Fare–like numerical data)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# -----------------------------
# Create a sample NYC Taxi Fare–like dataset
# Features: trip_distance (km), fare_amount ($)
# -----------------------------
np.random.seed(42)

# Normal trips
distance_normal = np.random.normal(loc=5, scale=2, size=450)
fare_normal = distance_normal * 2.5 + np.random.normal(0, 2, 450)

# Anomalous trips (very high fares or unusual distances)
distance_anomaly = np.random.uniform(20, 50, 50)
fare_anomaly = distance_anomaly * 6 + np.random.uniform(20, 60, 50)

# Combine data
distance = np.concatenate([distance_normal, distance_anomaly])
fare = np.concatenate([fare_normal, fare_anomaly])

df = pd.DataFrame({
    "trip_distance": distance,
    "fare_amount": fare
})

# -----------------------------
# Apply Isolation Forest
# -----------------------------
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.1,
    random_state=42
)

df["anomaly"] = iso_forest.fit_predict(df[["trip_distance", "fare_amount"]])
# -1 = anomaly, 1 = normal

# -----------------------------
# Visualize anomalies
# -----------------------------
plt.figure(figsize=(8, 6))

normal = df[df["anomaly"] == 1]
anomaly = df[df["anomaly"] == -1]

plt.scatter(
    normal["trip_distance"],
    normal["fare_amount"],
    label="Normal",
    alpha=0.6
)

plt.scatter(
    anomaly["trip_distance"],
    anomaly["fare_amount"],
    label="Anomaly",
    marker="x",
    s=80
)

plt.xlabel("Trip Distance (km)")
plt.ylabel("Fare Amount ($)")
plt.title("Isolation Forest – Taxi Fare Anomaly Detection")
plt.legend()
plt.show()
```

8. Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.


```python
# Train a SARIMA model on the AirPassengers dataset
# Forecast next 12 months and visualize results

import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# -----------------------------
# Load AirPassengers dataset
# -----------------------------
data = {
    "Month": pd.date_range(start="1949-01", periods=144, freq="M"),
    "Passengers": [
        112,118,132,129,121,135,148,148,136,119,104,118,
        115,126,141,135,125,149,170,170,158,133,114,140,
        145,150,178,163,172,178,199,199,184,162,146,166,
        171,180,193,181,183,218,230,242,209,191,172,194,
        196,196,236,235,229,243,264,272,237,211,180,201,
        204,188,235,227,234,264,302,293,259,229,203,229,
        242,233,267,269,270,315,364,347,312,274,237,278,
        284,277,317,313,318,374,413,405,355,306,271,306,
        315,301,356,348,355,422,465,467,404,347,305,336,
        340,318,362,348,363,435,491,505,404,359,310,337,
        360,342,406,396,420,472,548,559,463,407,362,405,
        417,391,419,461,472,535,622,606,508,461,390,432
    ]
}

df = pd.DataFrame(data)
df.set_index("Month", inplace=True)

# -----------------------------
# Train SARIMA model
# SARIMA(p,d,q)(P,D,Q,s)
# -----------------------------
model = SARIMAX(
    df["Passengers"],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 12)
)

results = model.fit(disp=False)

# -----------------------------
# Forecast next 12 months
# -----------------------------
forecast = results.get_forecast(steps=12)
forecast_mean = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# -----------------------------
# Plot results
# -----------------------------
plt.figure(figsize=(10, 5))

plt.plot(df.index, df["Passengers"], label="Observed")
plt.plot(forecast_mean.index, forecast_mean, label="Forecast")

plt.fill_between(
    forecast_mean.index,
    confidence_intervals.iloc[:, 0],
    confidence_intervals.iloc[:, 1],
    alpha=0.3
)

plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.title("SARIMA Forecast for AirPassengers (Next 12 Months)")
plt.legend()
plt.show()
```

9. Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.


```python
# Local Outlier Factor (LOF) for Anomaly Detection with Visualization

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# -----------------------------
# Create a numerical dataset
# Features: feature_1 and feature_2
# -----------------------------
np.random.seed(42)

# Normal data points
X_normal = np.random.normal(loc=0, scale=1, size=(400, 2))

# Anomalous data points
X_anomaly = np.random.uniform(low=-6, high=6, size=(40, 2))

# Combine dataset
X = np.vstack((X_normal, X_anomaly))
df = pd.DataFrame(X, columns=["feature_1", "feature_2"])

# -----------------------------
# Apply Local Outlier Factor
# -----------------------------
lof = LocalOutlierFactor(
    n_neighbors=20,
    contamination=0.1
)

df["anomaly"] = lof.fit_predict(df[["feature_1", "feature_2"]])
# -1 = anomaly, 1 = normal

# -----------------------------
# Visualize anomalies
# -----------------------------
plt.figure(figsize=(8, 6))

normal = df[df["anomaly"] == 1]
anomalies = df[df["anomaly"] == -1]

plt.scatter(
    normal["feature_1"],
    normal["feature_2"],
    label="Normal",
    alpha=0.6
)

plt.scatter(
    anomalies["feature_1"],
    anomalies["feature_2"],
    label="Anomaly",
    marker="x",
    s=80
)

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.legend()
plt.show()
```

2. Anomaly Detection in Streaming Data
Model Choice: Isolation Forest (Primary)

Why Isolation Forest?

- Fast and scalable for real-time systems
- Works well with high-frequency numerical data
- No need for labeled anomalies
- Suitable for sudden spikes or drops

Why not LOF or DBSCAN (primary)?
- LOF: Computationally expensive for streaming data
- DBSCAN: Sensitive to parameter tuning and struggles with evolving densities

Use Case
- Sudden consumption spikes (equipment failure, power theft)
- Sudden drops (sensor issues, outages)

Output interpretation

- Rows with -1 indicate abnormal spikes or drops

- These trigger alerts in real time

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Sample real-time like dataset
data = {
    "energy_usage": [320, 330, 340, 1000, 345, 350, 200, 360],
    "temperature": [30, 31, 32, 33, 31, 30, 29, 30]
}

df = pd.DataFrame(data)

# Train Isolation Forest
iso_forest = IsolationForest(
    n_estimators=100,
    contamination=0.1,
    random_state=42
)

df["anomaly"] = iso_forest.fit_predict(df)

# -1 = anomaly, 1 = normal
print(df)
```

3. Short-Term Energy Demand Forecasting
Model Choice: SARIMAX

Why SARIMAX?

- Energy demand has seasonality (daily, weekly patterns)
- External variables like temperature, humidity, region strongly affect demand
- SARIMAX supports exogenous variables
- Why not ARIMA or SARIMA alone?
- ARIMA ignores seasonality
- SARIMA ignores weather and region effects

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Sample time series
data = {
    "timestamp": pd.date_range("2024-01-01", periods=96, freq="15min"),
    "energy_usage": range(200, 296),
    "temperature": [30] * 96
}

df = pd.DataFrame(data)
df.set_index("timestamp", inplace=True)

# Train SARIMAX model
model = SARIMAX(
    df["energy_usage"],
    exog=df[["temperature"]],
    order=(1, 1, 1),
    seasonal_order=(1, 1, 1, 96)  # daily seasonality (96 intervals per day)
)

results = model.fit(disp=False)

# Forecast next 1 day (96 intervals)
forecast = results.get_forecast(
    steps=96,
    exog=df[["temperature"]].iloc[:96]
)

print(forecast.predicted_mean.head())
```
